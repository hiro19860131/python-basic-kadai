```python
# ================================
# 商品クラス（Product）
# 在庫管理で扱う「商品」という概念を表す設計図
# ================================
class Product:

# --------------------------------
# コンストラクタ（init）
# 商品が「新しく登録された瞬間」に実行される初期設定
# --------------------------------
  def init(self, name, stock, price):
    self.name = name # 商品名
    self.stock = stock # 初期在庫数
    self.price = price # 価格

# 初期化時にログを出す（動作が分かりやすい）
print(f"{self.name} を登録しました（在庫: {self.stock}, 価格: {self.price}円）")

# --------------------------------
# メソッド：在庫を追加する（入荷）
# --------------------------------
def add_stock(self, amount):
  self.stock += amount
  print(f"{self.name} を {amount} 個入荷しました（現在の在庫: {self.stock}）")

# --------------------------------
# メソッド：在庫を減らす（出荷）
# --------------------------------
def remove_stock(self, amount):
  self.stock -= amount
  print(f"{self.name} を {amount} 個出荷しました（現在の在庫: {self.stock}）")

# --------------------------------
# メソッド：現在の在庫数を返す（確認用）
# --------------------------------
def get_stock(self):
  return self.stock

# ================================
# ここから「呼び出し側」の処理
# 実際に商品を作って、メソッドを使って操作する部分
# ================================

# --- 商品Aを登録 ---
apple = Product("りんご", 30, 120)

# --- 商品Bを登録 ---
banana = Product("バナナ", 50, 80)

# --- りんごの在庫操作 ---
apple.add_stock(20) # 入荷
apple.remove_stock(10) # 出荷

# --- バナナの在庫操作 ---
banana.add_stock(15) # 入荷
banana.remove_stock(5) # 出荷

# --- 最終在庫の確認 ---
print(f"りんごの最終在庫: {apple.get_stock()}")
print(f"バナナの最終在庫: {banana.get_stock()}")
```

```python
from openpyxl import Workbook
from datetime import datetime

# 新しいExcelブックを作成
wb = Workbook()

# アクティブなシートを取得
ws = wb.active

# セルに値を入力
ws["B2"] = "請求書"
ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名：鈴木一郎 様"
ws["F4"] = "NO."
ws["F5"] = "日付"
ws["G4"] = "0001"

# A1に現在の日付を入力
ws["G5"] = datetime.today()

# 表示形式（例：2026/04/17）
ws["G5"].number_format = "yyyy/mm/dd"

# 列A-Gの幅を11に設定
for col in ["A", "B", "C", "D", "E", "F", "G"]:
    ws.column_dimensions[col].width = 11

# ヘッダー
headers = ["商品名", "数量", "単価", "金額"]

# ヘッダー入力（B10:E10）
for col, value in enumerate(headers, start=2):  # B列=2
    ws.cell(row=10, column=col, value=value)

# データ（商品名・数量・単価）
data = [
    ["りんご", 10, 100],
    ["みかん", 5, 200]
]

# データ入力（11行目以降）
for i, row_data in enumerate(data, start=11):
    ws.cell(row=i, column=2, value=row_data[0])  # 商品名（B列）
    ws.cell(row=i, column=3, value=row_data[1])  # 数量（C列）
    ws.cell(row=i, column=4, value=row_data[2])  # 単価（D列）

    # 金額（E列）に数式
    ws.cell(row=i, column=5, value=f"=C{i}*D{i}")

# ファイルを保存
wb.save("sample.xlsx")
```
