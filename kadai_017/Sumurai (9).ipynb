```python

class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # メソッドを定義する
    def check_adult(self):
        if self.age >=20: print("大人")
        else:
           print("大人でない")

# インスタンス化する
human1 = Human("ichiro", 10)
human2 = Human("taro", 25)
human3 = Human("jiro", 12)


# メソッドにアクセスして実行する
human1.check_adult()
human2.check_adult()
human3.check_adult()
```
