```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # メソッドを定義する
    def check_adult(self):
        if self.age >= 20:
            print(self.name, "は大人")
        else:
            print(self.name, "は大人でない")

# インスタンスをリストに格納
humans = [
    Human("ichiro", 10),
    Human("taro", 25),
    Human("jiro", 12)
]

# for文で順番に処理
for human in humans:
    human.check_adult()
```
