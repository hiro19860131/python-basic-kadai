```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    # メソッドを定義する
    def printinfo(self):
        self.name = "ichiro"
        self.age = "20"

        print(self.name)
        print(self.age)

# インスタンス化する
human = Human("name", "age")

# メソッドにアクセスして実行する
human.printinfo()
```
