```python
class Human:
    def __init__(self):
        self.name = ""
        self.age = ""
    # メソッドを定義する
    def printinfo(self, name, age):
        self.name = name
        self.age = age

        print(self.name)
        print(self.age)

# インスタンス化する
human = Human()

# メソッドにアクセスして実行する
human.printinfo("ichiro", 20)

```
